# A Retrospective Data Analysis of Anti-Mullerian Hormone Levels, Antral Follicle Count, and Oocyte Quantity in Intracytoplasmic Sperm Injection Patients 

**Introduction**
<br>Assisted Reproductive Technology (ART) involves treatments or procedures that includes the handling of human eggs and sperms or embryos for the purpose of establishing pregnancy. These complex procedures may be an option for people who have already gone through various infertility treatment options but who still have not achieved pregnancy. </br>
<br>Some ART options includes IUI, IVF and ICSI. In order to achieve successful pregnancy one has to go with the following steps:</br>

1.Sperm retrieval procedure</br>2.Artificial insemination</br> 3.Ovulation induction and oocyte retrieval</br> 4.Fertilisation through ICSI.</br> 

Infertility can be defined as not getting pregnant despite having carefully timed, unprotected sex for one year. The cause of infertility may be difficult to determine but may include inadequate levels of certain hormones in both men and women, and trouble with ovulation in women. Some of the major causes of infertility includes Ovulatory dysfunction (20%), Tubal disease (30%), Male infertility (30%) and Endometriosis (7%). Many treatments significantly improve the chances 
of getting pregnant. They include hormone treatments, fertility drugs and surgery. In addition, assisted reproduction uses various medical techniques to fertilise an egg.</br>

**ICSI – Intracytoplasmic sperm injection:**
<br>ICSI is an in vitro fertilisation procedure in which a single sperm cell is injected directly into the cytoplasm of an egg.</br> This technique is used in order to prepare the gametes for the obtention of embryos that may be transferred to a maternal uterus.</br> A special needle is used to go through the wall of the egg and the sperm is introduced into the egg’s cytoplasm after which the fertilisation takes place, division occurs and then the embryos are transferred into the women’s uterus.</br>

**ICSI differs from conventional IVF:**
<br>ICSI differs from conventional IVF in that the embryologists selects a single sperm to be injected directly into an egg, instead of fertilisation taking place in a dish where many sperms are placed near an egg.</br>

**ICSI indications:**
1. Severe problems in the seminal fluid as in severe deficiency in number or motility of the sperms or both.</br>
2. In cases when there are no sperms in the seminal fluid but there are in testicles(absent duct).In these cases the sperms are taken from either the epididymis or the testes (PESA,TESA)</br>
3. Problems with antibodies against sperm.</br>
4. Testicular cancer.</br>
5. Usage of frozen sperm in treatment which is not of optimum quality.</br>
**Stages of ICSI:**
<br>It includes,</br>
1. **Couple counselling** – All couples wishing to undergo ICSI should be counselled to the commencement of treatment. Counselling should address any concerns the couple may have also provide evidence based information regarding the likely outcomes and what to expect while undergoing ICSI. In addition, they should be aware of and discuss the increased risk of congenital bnormalities, particularly in male offspring.</br>
2. **Sperm collection** – Fresh ejaculate but this is not possible in all male partners contributing sperm for ICSI.</br>
3. **Course of hormone treatment** - A course of hormone treatment is given to stimulate her ovaries to produce several eggs. These women are then monitored using ultrasound.</br>
4. **Collection of eggs** – Once mature egg follicles have formed, the eggs are collected using a thin tube passed through the vagina. The eggs are then placed in special culture fluid in the laboratory.</br>
5. **Cleaning and dilution**- Sperm washing is a form of sperm preparation that is required is for IVF, because it removes chemicals from the semen, which may cause adverse reaction in the uterus, During the washing process, sperm is separated from the seminal fluid. This procedure enhances the fertilizing capacity of the sperm. </br>
6. **Intracytoplasmic sperm injection** – The mature egg is held with a holding pipette. A very sharp and hollow needle is used to immobilize and pick up a single sperm. This needle is then inserted through the zona and in to the center of the egg. The sperm is injected in the cytoplasm and the needle is removed. The eggs are checked the next 
morning for evidence of normal fertilization.</br>
**Benefits of ICSI:**
1. ICSI enables fertilization to happen when there are very few sperm available.</br>
2. Before the development of ICSI, male infertility related to abnormal semen count was considered largely untreatable. This assisted reproductive technique has therefore enable many couples who would not previously have been also to conceive, or who would have been reliant on donor sperm, to conceive, or who would have been reliant on donor sperm,to conceive a biological child.</br>
**AMH – Anti-Mullerian hormone:**</br>
* Anti-Müllerian hormone (AMH) is a dimericglycoprotein hormone from the transforming growth factor beta superfamily, who plays a key role in growth differentiation and folliculogenesis.</br>
* In humans, the gene for AMH is AMH and it is located on chromosome 19p13.3, while the gene AMHR2 codes for its receptor on chromosome 12. The two monomers constituting it are linked by disulphide bonds. </br>
* At the time of testicular differentiation to induce the regression of mullerian ducts, AMH is been produced by sertoli cells in male.In females, AMH is secreted by granulosa cells within preantral and small antral follicles up to 6 mm in diameter, to inhibit further recruitment of other follicles at folliculogenesis. </br>
* As a product of the granulosa cells, which envelop each egg and provide them energy, AMH can also serve as a molecular biomarker for relative size of the ovarian reserve.</br> 
* AMH can also be used as a marker for ovarian dysfunction, such as in women with polycystic ovary syndrome (PCOS).</br>
* At birth,AMH is hardly detected, then it peaks after puberty and declines with female aging reaching undetectable levels around menopause.</br>
* The normal AMH levels for IVF are 1.5 to 4 ng/ml, and women with such levels can get expect a good response rate to their fertility treatment and the level which is under 1.0 ng/ml is known to be a low AMH level and this indicates a very low ovarian reserve.</br>

**AFC – Antral follicle count:**
* An antral (resting) follicle is a small, fluid-filled sac that contains an immature egg. The follicles can be seen, measured and counted on Cycle Days 2, 3, and 5 by using ultrasound. The number of antral follicles varies from month to month.</br> 
* It measures a woman’s ovarian reserve or her remaining egg supply and reflects her fertility potential. </br>
* When an average to high number of antral follicles (eight or more) are visible on the ultrasound, fertility specialists expect to be able to retrieve a good number of eggs and the pregnancy rates are higher than average.</br> 
* If few antral follicles appear, a poorer response is expected and the IVF cycle could be cancelled to try for better results the following month. </br>
* Antral follicle count is rightly done on day 3 of the cycle by a Trans vaginal ultrasound.</br>
* Initially the ovarian volume of both the ovaries is calculated. Further the number of small antral follicles in both the ovaries is measured. These follicles could vary in size from 2-10 mm.</br>

<br>**Rationale**</br>

In order to attain successful ART for infertility patients, there is a need to identify women with a high risk of producing poor response to stimulation and who have a very low probability of becoming pregnant as well as those who still have the ability to produce enough oocytes with good quality even if her age is advanced.</br>

The serum AMH level is a reliable biomarker for ovarian reserve (the number of oocytes present). By measuring the AMH level we can predict the chances of successful pregnancy. Also, AMH serves as a diagnostic and prognostic tool for PCOS (Polycystic ovarian syndrome) patients.</br>

The AFC count is also major predictor of the ovarian reserve and could reflect the size of the remaining primordial follicle pool. Low fertility is characterized by low numbers of oocytes and are usually associated with high follicle stimulating hormone (FSH) levels.</br>

To maintain a patient’s confident level and satisfaction, creating expectation is very important. These tests which are performed in this study would help the clinicians have better predictions and they can make adjustments in the gonadotropin dosage thereby preventing ovarian hyper stimulation syndrome. These predictions may also supplement the presently available tests of ovarian reserve.</br>


**Problem Statement**</br>
The prime objective of this project is to perform a retrospective analysis using bio statistical tools and determine the relation between Anti mullerian hormone level, Antral Follicle count and the quantity of oocytes in ICSI patients. Also,  to determine the relationship between Anti mullerian hormone level and Polycystic Ovarian Syndrome.


<br>**Data Collection**</br>This retrospective study includes 97 women aged between 20-45 years old referred to ICSI 
treatment in Vamsam infertility clinic, Shree Shiva Hospital, Coimbatore, Tamil Nadu from 
January 2016 to December 2018. Data collected from patient’s medical record included age, 
weight, AMH, FSH, E2, LH, AFC and clinical record included causes of infertility, number of 
previous pregnancies, number of oocytes punctured, number of oocytes retrieved, number of 
oocytes fertilized, pregnancy outcome and cancellation of cycle.

Since, most columns contain outliers, lets choose IQR value imputation to treat/remove the outliers.

```python
# create a new dataframe with numeric attributes
numeric_data = data.select_dtypes(include = np.number)

# find the Q1, Q3 and IQR values for all columns
Q1 = numeric_data.quantile(0.25)
Q3 = numeric_data.quantile(0.75)
IQR = Q3-Q1

print("The interquartile Range values of the dataset :")
print(IQR)
```

```python
# create a new dataframe with categoric attributes
categoric_data = data.select_dtypes(include = 'object')

# display the data
categoric_data.head(10)
```

```python
# check the number of rows with outliers before capping
outliers = numeric_data[((numeric_data < (Q1 - 1.5 * IQR)) |(numeric_data > (Q3 + 1.5 * IQR))).any(axis=1)] 

outliers.shape
```

```python
# Replace every outlier on the lower side by the lower whisker
for i, j in zip(np.where(numeric_data < Q1 - 1.5 * IQR)[0], np.where(numeric_data < Q1 - 1.5 * IQR)[1]): 
    
    lwhisker  = Q1 - 1.5 * IQR #lower whisker
    numeric_data.iloc[i,j] = lwhisker[j]
    
    
#Replace every outlier on the upper side by the upper whisker    
for i, j in zip(np.where(numeric_data > Q3 + 1.5 * IQR)[0], np.where(numeric_data > Q3 + 1.5 * IQR)[1]):
    
    uwhisker  = Q3 + 1.5 * IQR #upper whisker
    numeric_data.iloc[i,j] = uwhisker[j]
    
```

```python
# check the number of rows with outliers after capping
outliers_capping = numeric_data[((numeric_data < (Q1 - 1.5 * IQR)) |(numeric_data > (Q3 + 1.5 * IQR))).any(axis=1)] 

outliers_capping.shape
```

```python
# create bin edges 
bin_edges_amh = [0,2,6,final_data['AMH'].max()]
bin_edges_afc = [0,6,11,final_data['AFC'].max()]

# create labels
label_amh = ['Low','Medium','High']
label_afc = ['Poor', 'Normal', 'High']

# create new attributes based on categories
final_data['AMH_group'] = pd.cut(final_data['AMH'], bins = bin_edges_amh, labels = label_amh)
final_data['AFC_group'] = pd.cut(final_data['AFC'], bins = bin_edges_afc, labels = label_afc)

# display data
final_data.head(10)
```

```python
# create a new feature causes_grouping
def grouping(row):
    if (row['Female_Causes'] != 'Not Applicable') and (row['Male_Causes'] != 'Not Applicable'):
        return 'Both'
    elif (row['Female_Causes'] != 'Not Applicable') and (row['Male_Causes'] == 'Not Applicable'):
        return 'Female'
    elif (row['Female_Causes'] == 'Not Applicable') and (row['Male_Causes'] != 'Not Applicable'):
        return 'Male'
    else:
        return 'None'

final_data['Causes_grouping'] = final_data.apply(grouping, axis=1)


# display the data
final_data.head(10)
```

```python
# Assumption to be checked:
#  1. check for normality of the data
# hypothesis
H0 = 'Data is normal'
H1 = 'Data is not normal'

shapiro_stat, shapiro_p_value = stats.shapiro(final_data['Age'])
print("\nShapiro test statistics :", shapiro_stat)
print("\nshaprio p-value:", shapiro_p_value)

if pvalue_amh < 0.05:
    print(f'\n{H1} as the p_value ({pvalue_amh.round(3)}) < 0.05')
else:
    print(f'\n{H0} as the p_value ({pvalue_amh.round(3)}) > 0.05')
    
# create groups
low_amh = final_data[final_data['AMH_group'] == 'Low']['Age']
medium_amh = final_data[final_data['AMH_group'] == 'Medium']['Age']
high_amh = final_data[final_data['AMH_group'] == 'High']['Age']


# 2. check for equality of variance
# Hypothesis
H0 = 'Data has equal variance'
H1 = 'Data has unequal variance'

levene_stat, levene_pvalue = stats.levene(low_amh, medium_amh, high_amh)
print("\nLevene test statistics :", levene_stat)
print("\nLevene p-value :", levene_pvalue)

if pvalue_amh < 0.05:
    print(f'\n{H1} as the p_value ({pvalue_amh.round(3)}) < 0.05')
else:
    print(f'\n{H0} as the p_value ({pvalue_amh.round(3)}) > 0.05')
```

**One way ANOVA to check if age has an effect on AMH Level**</br>

**Chisquare test to check if Pregnancy outcome is influenced by AMH level.**</br>
Null Hypothesis(H0) : AMH level has no effect on pregnancy outcome.</br>
Alternate Hypothesis(H1) : AMH level has an effect on pregnancy outcome.</br>
    
`Significance Level = 0.05`

```python
# hypothesis formulation
H0 = "AMH level has no effect on pregnancy outcome"   # Stating the Null Hypothesis
H1 = "AMH level has an effect on pregnancy outcome"   # Stating the Alternate Hypothesis

crosstab_amh = pd.crosstab(final_data['AMH_group'], final_data['Pregnancy outcome'])  # Contingency table 

chi_amh, pvalue_amh, dof_amh, expected_amh =  stats.chi2_contingency(crosstab_amh)

if pvalue_amh < 0.05:  
    print(f'{H1} as the p_value ({pvalue_amh.round(3)}) < 0.05')
else:
    print(f'{H0} as the p_value ({pvalue_amh.round(3)}) > 0.05')
```

**Failing to reject the null hypothesis. Thus, Pregnancy outcome is irrespective of AMH Level.**
